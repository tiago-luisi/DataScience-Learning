### Exercício 1: Gráfico de Uma Função Linear

Crie um gráfico para a função y = 2x + 3 para valores de x no intervalo de 5 a 10.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
  y = []
  for i in range(len(x)):
    y.append(2*x[i] + 3)
  return y
```

```python
#y = 2x + 3
x=[]
x = np.linspace(5,10,100)
```

```python
y = func(x)
```

```python
plt.plot(x,y, label='y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da Função Linear - Intervalo de 5 a 10 ')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 2: Gráfico de uma Função Quadrática

Desenhe o gráfico da função y = x^2 - 4x + 4. Crie uma massa de dados de exemplo usando np.linspace().

```python
def func2(x):
  y = []
  for i in range(len(x)):
    y.append((x[i]**2) - (4*x[i]) + 4)
  return y
```

```python
#y = 2x + 3
x=[]
x = np.linspace(-2,6,100)
```

```python
y = func2(x)
```

```python
plt.plot(x,y, label='y = x^2 - 4x + 4')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da Função Quadratica')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 6: Resolver uma Equação Quadrática

Escreva um script em Python que resolve a equação x^2 − 5x + 6 = 0 e imprime as raízes.

```python
#Equação Quadratica se resolve com o Delta
#formula Bhaskara.
import cmath

a, b, c= 1, -5, 6
delta = (b**2) - (4*a*c)
sol1 = (-b-cmath.sqrt(delta))/(2*a)
sol2 = (-b+cmath.sqrt(delta))/(2*a)

print(sol1,sol2)


```
